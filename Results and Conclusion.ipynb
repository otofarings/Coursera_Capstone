# Results and Conclusion

In the course of the work, an analysis was made of apartments in the city of Penza, which are rented. Enough information was received to answer the task, which was to obtain additional information about the area in which these apartments are located. This information will help you navigate the choice.

For analysis, we processed the information using Python libraries, and received additional data using the Foursquere API. For clarity, the data has been displayed on a map.


Further selection is made by the potential renter of the apartment, based on personal preferences. For example: if a person wants to rent an apartment in a calm and quieter area, then you should choose clusters in which bars and clubs are not in the top 10 popular places. In our case, in the city of Penza, this is the third cluster.

If a person goes in for sports, then it would be right to find an apartment in a cluster where the gym is in the top 10 places. Etc.

After choosing the region, you can continue with the standard selection by price and other parameters. There are enough of them in each district, since the clusters correlate weakly with the parameters of the apartment.
